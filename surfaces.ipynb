```
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```
x_interval = (-10, 10)
y_interval = (-10, 10)
x_points = np.linspace(x_interval[0], x_interval[1], 100)
y_points = np.linspace(y_interval[0], y_interval[1], 100)
X, Y = np.meshgrid(x_points, y_points)
```

```
def func3d(x, y):
    return np.cos(x) * np.cos(y)
func3d_vec = np.vectorize(func3d) # Vectorisation de la fonction pour accélérer
# l'exécution.
Z = func3d_vec(X, Y)
```

```
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title='My function', autosize=True,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
```
